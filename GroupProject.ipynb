```python
# First Graph of the maps
G = nx.Graph()
labels = {}
for i in range(0,22):
  labels[i] = i
G.add_nodes_from(nodes)
weight = []
for i,j in edges: 
  x1,x2 = stop_pos[i]
  y1,y2 = stop_pos[j]
  weight_v = np.sqrt((y1-x1)**2+(y2-x2)**2)
  weight.append(weight_v)
for i in range(len(edges)):
  G.add_edges_from([edges[i]],weight = weight[i])
nx.draw_networkx_labels(G,pos=stop_pos, labels=labels)
nx.draw(G,stop_pos)

#delete one node that have many edges, see what happened, more efficent or not?(If we have time think about this) (part 2)
```

```python
# Find the shortest path length for the shortest path for all pairs of nodes in the Graph 
import matplotlib.pyplot as plt
p = nx.shortest_path_length(G)
dict_p = dict(p)
list_p = []
for i in range(0,22):
    for j in range(0,22):
        list_p.append(dict_p[i][j])
plt.hist(list_p)
plt.xlabel('shortest length')
plt.ylabel('count')
```

```python
#find bus station
deg = G.degree()
count = [d for n,d in deg]
plt.hist(count)
plt.xlabel('degree')
plt.ylabel('count')
deg
```
